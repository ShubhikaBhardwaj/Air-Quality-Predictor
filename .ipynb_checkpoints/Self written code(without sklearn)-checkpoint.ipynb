```python
def hypothesis(x,theta):
    #return theta[1]*x + theta[0]
    return np.dot(x,theta)
    
def gradient(x,y,theta):
     
    grad = np.zeros(x.shape[1])
    m = x.shape[0]
    thetas=[]
    
    
    
    for i in range(x.shape[1]):
        hx= hypothesis(x[i],theta)    
        for j in range(m):
            
            grad[i] += 2*(hx-y[j])*x[j][i]
            thetas.append(grad[i]/m)
            

    return thetas

#for i in range(m):
#        hx = hypothesis(x[i],theta)
#        grad[1] += (hx-y[i])*x[i]
#        grad[0] += (hx-y[i])

 #   return grad/m
#


def error(x,y,theta):
    
    e = 0
    for i in range(x.shape[0]):
        e += (y[i] - hypothesis(x[i],theta))**2
    
    #print(type(e))
    return 0.5*e/x.shape[0]

def gradientDescent(x,y,lr=.01, threshold=0.001):
    theta = np.array([0.1,0.2,0.1,0.1,0.1])
    error_list = []
    # Homework - define converge criteria to break the loop
    # Change in error < threshold
    
    for i in range(800): 
        grad = gradient(x,y,theta)
        e = error(x,y,theta)
        error_list.append(e)
        #upperlimit=5(val)
        val=x.shape[1]
        for j in range(val):
            theta[j]=theta[j]-lr*grad[j]
        
    return theta,error_list
```

```python
theta,err= gradientDescent(x,y)
```
